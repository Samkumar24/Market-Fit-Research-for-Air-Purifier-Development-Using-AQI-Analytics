### Data Cleaning Note:

The `status` column indicates whether an outbreak was "reported in same week."  
However, if the number of `days` between `outbreak_starting_date` and `reporting_date` is more than 7,  
the outbreak should be treated as **"reported late"** instead.  
We corrected such mismatches to ensure data consistency in timeliness reporting.


```python
# Calculates how many days passed between the outbreak start and when it was reported

illnes_df['days'] = (pd.to_datetime(illnes_df['reporting_date']) - pd.to_datetime(illnes_df['outbreak_starting_date'])).dt.days
illnes_df.sample(5)
```

```python
# Create a new column 'status_2' based on conditions:
# - If 'status' is 'previous week follow up' ‚Üí assign same to 'status_2'
# - Else, if 'days' <= 7 ‚Üí label as 'reported same week'
# - Else ‚Üí label as 'reported late'
illnes_df['status_2'] = np.where(
    illnes_df['status'] == 'previous week follow up',
    'previous week follow up',
    np.where(illnes_df['days'] <= 7, 'reported same week', 'reported late')
)
```

### Using Median Value to fill the missing the Dates 

We use the **median** to fill missing values in the `days` column because the data is **right-skewed** with potential outliers. The **mean** could be distorted by these extreme values, while the **median** provides a more accurate and robust central value for imputation.


```python
# Plots a boxplot of 'days' for cases that were reported late and have a valid reporting date

illnes_df[(illnes_df['status_2'] == 'reported late') &(illnes_df['reporting_date'].notnull())].days.plot.box()
```

#### This boxplot shows how long it took to report cases that were marked as 'reported late'.

- Most were reported within a short time, but a few took much longer than usual, shown as outliers.


```python
# Calculate the median of 'days' for rows where:
# - 'status_2' is 'reported late'
# - 'reporting_date' is not null
median_fill_value = illnes_df[
    (illnes_df['status_2'] == 'reported late') & 
    (illnes_df['reporting_date'].notnull())
].days.median()

# Display the calculated median value
median_fill_value
```

```python
# Define a function to determine the correct 'reporting_date'
def reporting_date(df):
    return np.where(
        # If status is 'previous week follow up', use 'outbreak_starting_date'
        df['status'] == 'previous week follow up',
        df['outbreak_starting_date'],
        
        # Else, if 'reporting_date' is missing, fill with outbreak_starting_date + 31 days
        np.where(df['reporting_date'].isnull(),
            df['outbreak_starting_date'] + pd.to_timedelta(31, 'd'),
            
            # Else, keep the existing 'reporting_date'
            df['reporting_date']
        )
    )

# Apply the function to update the 'reporting_date' column
illnes_df['reporting_date'] = reporting_date(illnes_df)
```

### Key Air Pollution Control Policies & Their Impact

- üîµ **National Clean Air Programme (NCAP) (2019)**  
  Launched to reduce particulate matter (PM) pollution by up to 40% in 131 cities with poor air quality. Cities like Mumbai and Kolkata have shown significant improvements, with PM‚ÇÅ‚ÇÄ levels dropping by 44% and 37%, respectively. However, progress has been uneven‚Äîcities like Delhi and Faridabad have struggled due to underutilized budgets and challenges in implementation.







- üìä **SAFAR (System of Air Quality Forecasting & Research)**  
  Provides real-time air quality data and short-term forecasts (1 to 3 days) in metros such as Delhi, Mumbai, Pune, and Ahmedabad. This enables authorities and citizens to take timely preventive actions, improving public health and lowering medical costs.
  
  
  
  
  
  

- üöó **Bharat Stage VI (BS-VI) Emission Norms (2020)**  
  A major upgrade in vehicle emission standards that reduces harmful pollutants like nitrogen oxides (NO‚Çì) and particulate matter (PM), aligning India‚Äôs vehicle standards with global benchmarks.






- ‚ö° **Petrol and Diesel Vehicle Ban Proposal in Mumbai**  
  Maharashtra is considering banning petrol and diesel vehicles in Mumbai to replace them with electric or cleaner alternatives, supporting better air quality goals.





- üè≠ **Emission Control Challenges in Maharashtra‚Äôs Coal Power Plants**  
  Many coal-fired power plants received exemptions from installing pollution control devices like Flue Gas Desulfurization (FGD) systems as of July 2025. This risks increased PM‚ÇÇ.‚ÇÖ and SO‚ÇÇ emissions, undermining air quality improvements.
